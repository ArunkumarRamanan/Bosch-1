Data scraped from kaggle page using google sheets:

```r
ranks = read.csv("/Users/alexpapiu/Documents/Kaggle/Bosch/Rankings/scores.csv")
```

```r
head(ranks)
```

```r
ggplot(ranks, aes(x = Score)) +
    geom_histogram(color = "black", bins = 20) +
    geom_vline(xintercept = 0.26912, color = "red") +
    ggtitle("Rankings - Red Line is Us")
```

```r
ranks %>%
    filter(Score > 0.25) %>%
    ggplot(aes(x = Score)) +
    geom_histogram(color = "black", bins = 15) +
    geom_vline(xintercept = 0.26912, color = "red")
```
